```python
def linkdin_scrap(url_list,xpath_dict):
    url_list = url_list
    xpath_dict = xpath_dict
    gender_data ="//main/div//section//artdeco-tabs/artdeco-tabpanel[1]/div//div/blockquote/div"
    p_company_x2 ="/html/body/div//main/div//span/div/section/div[1]/section/ul/li[2]/section/div/a/div//h3/span[2]"
    p_title_x2 = "/html/body/div//main/div//span/div/section/div[1]/section/ul/li[2]/section/ul/li[1]/div//h3/span[2]"
    column_s = [['url','gender']+[*xpath_dict.keys()]][0]
    final_df = pd.DataFrame(columns =column_s)
    s_data =[]
    #___text_scraper
    def xpath_s(xpath):
        data_list = []
        data_x = driver.find_elements_by_xpath(xpath)
        if data_x == []:
            data_list.append("No Data found")

        else:
            for i in data_x:
                data_list.append(i.text.replace('\n',' ').replace('\t',' '))
                
        return data_list
    
    #___xpaths

    for url in url_list:
        sdata = driver.get(url)
        data_dic ={}
        data_dic['url'] = url
        scheight = 2
        while scheight > 0.1:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight/%s);" % scheight)
            scheight -= 0.01

        try:
            g_data = driver.find_elements_by_xpath(gender_data)[0].text.lower()
            print(g_data)

            if (' her ' in re.findall(' her |she ',g_data,re.IGNORECASE) or 'she ' in re.findall(' her |she ',g_data,re.IGNORECASE)):
                data_dic['gender'] = 'female'

            elif (' his ' in re.findall(' his |he ',g_data,re.IGNORECASE) or 'he ' in re.findall(' his |he ',g_data,re.IGNORECASE)):
                data_dic['gender'] = 'male'

            else:
                data_dic['gender'] = 'unknown'
        except:
            data_dic['gender'] = 'unknown'
        
        print(data_dic['gender'])
        for i in [*xpath_dict]:
            data_dic[i] = xpath_s(xpath_dict[i])
            print(data_dic[i])
            
        if data_dic['previous_title'] == ['No Data found']:
            data_dic['previous_title'] = xpath_s(p_title_x2)
            
        if data_dic['previous_company'] == ['No Data found']:
            data_dic['previous_company'] = xpath_s(p_company_x2)

        data_df = pd.DataFrame(data_dic)
        final_df = final_df.append(data_df, ignore_index = True)
#         final_df.to_csv('output/output_in_progress_'+today+'.csv',sep='\t', encoding='utf-8', index=False)
        

        
#     Linkdin_data = pd.DataFrame(s_data,columns=['URL', 'Name', 'Title', 'Location','current_company'])
    return final_df
        
```
