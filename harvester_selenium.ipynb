```python
company = driver.find_elements_by_xpath("//a[@class='listing-name']")
company_data = []
for j in company:
    company_data.append(j.text)

company_data = list(dict.fromkeys(company_data))

row = []
for i in company_data:
    match = r""+i+""
    xtadd = "//table[contains(.,"+"\""+match+"\""+")]/../../../..//*[@class='listing-heading']"
    type_address = driver.find_elements_by_xpath(xtadd)
    if type_address == []:
        xtadd2 = "//a[@class='listing-name'][contains(.,"+"\""+match+"\""+")]/../../../../../..//*[@class='listing-heading']"
        type_address2 = driver.find_elements_by_xpath(xtadd2)
        if type_address2 == []:
            tadd = "no data found"
            
        else:
            for j in type_address2:
                tadd = j.text
        
    else:
        for j in type_address:
            tadd = j.text
        
    xdesc = "//table[contains(.,"+"\""+match+"\""+")]/../../../..//*[@class='listing-short-description']"
    desc = driver.find_elements_by_xpath(xdesc)
    if desc == []:
        xdesc2 = "//a[@class='listing-name'][contains(.,"+"\""+match+"\""+")]/../../../../../..//*[@class='listing-short-description']"
        desc2 = driver.find_elements_by_xpath(xdesc2)
        if desc2 == []:
            desc_d = "no data found"
        else:
            for j in desc2:
                desc_d = j.text        
    else:
        for j in desc:
            desc_d = j.text
            
    xphon = "//table[contains(.,"+"\""+match+"\""+")]/../../../..//*[@title = 'Phone' ]"
    phone = driver.find_elements_by_xpath(xphon)
    if phone == []:
        xphon2 = "//a[@class='listing-name'][contains(.,"+"\""+match+"\""+")]/../../../../../..//*[@title = 'Phone' ]"
        phone2 = driver.find_elements_by_xpath(xphon2)
        if phone2 == []:
            phone_d = "no data found"
        else:
            for j in phone2:
                phone_d = j.get_attribute('href')        
    else:
        for j in phone:
            phone_d = j.get_attribute('href')
        
    xemail = "//table[contains(.,"+"\""+match+"\""+")]/../../../..//*[@class = 'contact contact-main contact-email ']"
    email = driver.find_elements_by_xpath(xemail)
    if email == []:
        xemail2 = "//a[@class='listing-name'][contains(.,"+"\""+match+"\""+")]/../../../../../..//*[@class = 'contact contact-main contact-email ']"
        email2 = driver.find_elements_by_xpath(xemail2)
        if email2 == []:
            email_d = "no data found"
        else:
            for j in email2:
                email_d = j.get_attribute('data-email')        
    else:
        for j in email:
            email_d = j.get_attribute('data-email')
            
    xurl = "//table[contains(.,"+"\""+match+"\""+")]/../../../..//*[@class= 'contact contact-main contact-url ' ]"
    urle = driver.find_elements_by_xpath(xurl)
    if urle == []:
        xurl2 = "//a[@class='listing-name'][contains(.,"+"\""+match+"\""+")]/../../../../../..//*[@class= 'contact contact-main contact-url ' ]"
        urle2 = driver.find_elements_by_xpath(xurl2)
        if urle2 == []:
            url_d = "no data found"
        else:
            for j in urle2:
                url_d = j.get_attribute('href')        
    else:
        for j in urle:
            url_d = j.get_attribute('href')
            
    row.append([i, tadd, desc_d, phone_d, email_d, url_d])
    
dataf = pd.DataFrame(row,columns=['company', 'type_address', 'desc', 'phone', 'email', 'url'])
```

```python
def linkdin_scrap(url_list):
    url_list = url_list
    s_data =[]
    #___text_scraper
    def xpath_s(xpath):
        data_list = []
        data_x = driver.find_elements_by_xpath(xpath)
        if data_x == []:
            data_list.append("No Data found")

        else:
            for i in data_x:
                data_list.append(i.text)
                
        return data_list
    
    #___xpaths
    lk_name = "//ul/li[contains(@class,'t-24')]"
    lk_title = "//h2[contains(@class,'mt1')]"
    lk_location = "//ul/li[contains(@class,'t-16')]"
    lk_Current_company ="//span[contains(@class,'t-14')]"
    for url in url_list:
        sdata = driver.get(url)
        lkurl = url
        print(lkurl)
        lkname = xpath_s(lk_name)
        lktitle = xpath_s(lk_title)
        lklocation = xpath_s(lk_location)
        combined = [url, lkname[0], lktitle[0], lklocation[0]]
        print(combined)
        s_data.append(combined)
        
    Linkdin_data = pd.DataFrame(s_data,columns=['URL', 'Name', 'Title', 'Location','current_company'])
    return Linkdin_data
        
```

```python
list_urls = pd.read_csv('LinkedIn_Google_Filtered.csv', sep ='\t')['url'].values
```
