```python
def linkdin_url_generator(companies):
    base_url = 'https://www.google.com/search?q='
    position_list = ['merchendise','director','ceo','manager']
    Data_per_page = 30
    l_url_list = []
    companies = companies
    for i in companies:
        for j in position_list:
            l_url_list.append(base_url+'site:linkedin.com/in/+'+i+'+'+j+'&num='+str(Data_per_page))
            
    return l_url_list
```
